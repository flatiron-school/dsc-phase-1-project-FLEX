```python
# Import standard packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pd.set_option('display.float_format', lambda x: f'{x:,.2f}' )
```

```python
# Here you run your code to explore the data
import glob, os
fpath = 'zippedData/'
os.listdir(fpath)
```

```python
query = fpath+"*.gz"

file_list = glob.glob(query)
file_list
```

```python
tables = {}
for file in file_list:
    print('---'*20)
    file_name = file.replace('zippedData/','').replace('.','_')

    print(file_name)
    
    if '.tsv.gz' in file:
        temp_df = pd.read_csv(file, sep="\t",encoding="latin-1")
    else:
        temp_df = pd.read_csv(file)
    display(temp_df.head(),temp_df.tail())
    

    tables[file_name] = temp_df
```

```python
bom_movie_gross = tables['bom_movie_gross_csv_gz'].copy()
bom_movie_gross.head(3)
```

```python
imdb_title_basics = tables['imdb_title_basics_csv_gz'].copy()
imdb_title_basics.head(10)
```

```python
df = pd.merge(bom_movie_gross,imdb_title_basics,left_on='title',
         right_on='original_title')
df
```

```python
df[~df.duplicated(keep=False)]
```

```python
## Examining duplicates according to those 3 cols
df[df.duplicated(keep=False,subset=['title','original_title','domestic_gross'])]
```

- Noticed there are sseveral duplciates for many movies. They are only duplicates accoridng to some columns.
- We are going to keep = 'first'

```python
## Dropping duplicaters  & verifying 
df = df.drop_duplicates(keep='first',subset=['title','original_title','domestic_gross'])
df[df.duplicated(keep=False,subset=['title','original_title','domestic_gross'])]
```

```python
df
```

### Dealing with Genres

```python
ex = df.loc[0,'genres']
ex
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['genres'])
df
```

```python
ex.split(',')
```

```python
df['genres_list'] = df['genres'].apply(lambda x: x.split(','))
df
```

```python
genre_df = df.explode('genres_list')
genre_df
```

```python
sns.regplot(data=df,x='runtime_minutes',y='domestic_gross',line_kws={'color':'black'})
```

```python
sns.lmplot(data=genre_df,x='runtime_minutes',y='domestic_gross',
           col='genres_list',hue='genres_list',col_wrap=3,line_kws={'color':'black'})
```

```python
df['genres_list'].map(lambda x: len(x)).value_counts()
```

```python
fig,ax = plt.subplots(figsize=(8,4))
sns.barplot(data=genre_df,x='genres_list',y='domestic_gross',ax=ax)
ax.set_xticklabels( ax.get_xticklabels(),rotation=45,ha='right');
```
